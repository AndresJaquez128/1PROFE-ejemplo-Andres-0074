```python
class Automovil:
  def __init__(self, marca, placa, precio):
    self.marca = marca
    self.placa = placa
    self.precio = precio

  def leerdatos(self):
    self.marca = input("Ingrese la marca: ")
    self.placa = input("Ingrese la placa: ")
    self.precio = input("Ingrese el precio: ")
    return self.marca, self.placa, self.precio

  def mostrardatos(self):
    print("Marca: ", self.marca)
    print("Placa: ", self.placa)
    print("Precio: ", self.precio)
    return self.marca, self.placa, self.precio
# fin de la clase automovil

# creando los objetos carro1, carro2, carro3
# Se deben crear instancias separadas para cada carro
print("Carro Numero 1")
# Llama a los métodos en la instancia del objeto, no en la clase
carro1 = Automovil("", "", "") # creando el objeto carro1
carro1.leerdatos()
carro1.mostrardatos()
print(60 * "=")

print("Carro Numero 2")
carro2 = Automovil("", "", "") # creando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print(60 * "=")

print("Carro Numero 3")
carro3 = Automovil("", "", "") # creando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print(60 * "=")
print("Software desarrollado por Jaquez Andres NL = 27")
```
